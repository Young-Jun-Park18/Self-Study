```python
#if문의 성질이에요.
weather = input("오늘 날씨는 어떤가요? : ")
if weather == "비":
  print("우산을 챙겨주세요\n")
elif weather == "태풍":
  print("나가지마시오\n")
elif weather == "맑음":
  print("양산을 준비해주세요\n")
else:
  print("준비물 없음")
temp = int(input("오늘은 몇도인가요? ")) #입력받는 값이 숫자이기 때문에 int()로 묶어주지 않으면 문자로 인식한다.
if 30 <= temp:
  print("폭염주의\n")
elif temp >= 10 and temp < 30:
  print("적당한 날씨입니다\n")
else:
  print("겉옷을 챙기세요\n")

  
#여기부터는 반복문입니다.
for wait_num in range(1,4):
  print("대기번호 : {}".format(wait_num))
mega_coffe = ["박영준","고선영","김진우"]
for customer in mega_coffe:
  print("{}님, 주문하신 커피가 나왔습니다".format(customer))
#for 새로운 변수 in 줄 변수, .format() 자주 쓰임
customer = "박영준"
index = 5
while index >= 1:
  print("{0}님, 주문하신 커피 나왔습니다. {1}번 반복할 것 같습니다.".format(customer, index))
  index -= 1
  if index == 0:
    print("매진입니다.")
# {a}{b}.format(a,b)를 해야 순서에 알맞게 대입
customer2 = "박영준"
index2 = 1
while False:
  print("{0}님, 커피 안 가져가면 계속 호출합니다.{1}회 불렀습니다. 당장 나오세요".format(customer2, index2))
  index2 += 1


customer3 = "박영준"
person = "Unknown"
while person != customer3:
  print("{0}님, 주문하신 커피가 나왔습니다.".format(customer3))
  person = input("성함이 어떻게 되십니까? ")
  if person == "박영준":
    print("맛있게 드십시오.")    
  else:
    print("당신의 커피가 아닙니다 정신 차리세요")

absent = [2, 5] #결석 학생 출석번호
no_books = [8]
for student in range(1,11):
  if student in absent:
    print("이 학생은 결석입니다.")
    continue
  elif student in no_books:
    print("{}번, 수업 끝나고 교무실로 오세요".format(student)) 
    continue    
  print("출석번호 {}번, 일어나서 본문 해석해봐".format(student))
#for문과 마찬가지로 if문에도 in을 사용할 수 있다
#continue를 사용해서 특정 조건을 건너뛸 수 있다.

student2 = list(range(1, 101))
print(student2)
student2 = [i+100 for i in student2]
print(student2)
#한줄 for문은 대괄호를 먼저 써주고 for 변수 in 변수 앞에 추가 조건을 붙이면 된다


student4 = ["young jun", "sun young", "jin woo", "yun woo"]
student4 = [i.capitalize() for i in student4]  # capitalize()가 앞글자만 대문자로 바꿔준다
print(student4)

#QUIZ 당신은 Cocoa 서비스를 이용하는 택시기사입니다. 50명의 승객과 매칭 기회가 있을 때, 총 탑승 승객 수를 구하는 프로그램을 작성하시오.__doc__

#조건1 : 승객별 운행 소요 시간은 5~50분 사이의 난수로 정해집니다.
#조건2 : 당신은 소요 시간 5분 ~ 15분 사이의 승객만 매칭해야 합니다.

#(출력문 예제)
#[0] 1번째 손님 (소요시간 : 15분)
#[ ] 2번째 손님 (소요시간 : 50분)
#[0] 3번째 손님 (소요시간 : 5분)
#...
#[ ] 50번째 손님 (소요시간 : 16분)
#총 탑승 승객 : 2명
import random #랜덤 모듈을 import로 추가
i = 0
count = 0
customer = list(range(1, 51)) #승객을 50명으로 설정
for time in customer:
  time = random.randrange(5, 51) #1~51난수 생성
  if time >= 5 and time <= 15: 
    possible = "o"
    count += 1
  else:
    possible = " "
  print("[{0}] {1}번째 손님 (소요시간 : {2}분)".format(possible, customer[i], time))  
  i += 1
print("총 탑승 승객 : {0}명".format(count))
  
    
  
#첫 번째 오류 수정 : if문 안 변수를 customer[i]로 했더니 손님 번호가 5~15번인 사람만 앞에 동그라미가 쳐졌음
#두 번쨰 오류 수정 : 화면에 n번째 손님이 출력되지 않고 customer[i]번 째 손님이라고 출력됨 --> 해결 방법 : ""따옴표 안에 customer[i]라고 쓰지 않고 수정할 자리에 {}를 사용한 뒤 뒤에 .format(customer[i]라고 바꿈) 

#궁금한점 : 아직 if + in 의 정확한 의미를 모르겠음
#궁금한점2: .format은 형식에 구애받지 않고 사용할 수 있는가?
  
```
