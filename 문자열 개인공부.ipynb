# 문자열 개인공부

```python
##슬라이싱
my_number = "990629-1234567"
print("성별 : " + my_number[7]) #대괄호로 부분 발췌가능
print("생년 : " + my_number[0:2])#콜론으로 범위 발췌가능
print("월 : " + my_number[2:4])
print("일 : ", my_number[4:6])
print("생년월일 : ", my_number[:6]) #시작부분은 생략가능
print("뒤 7자리 : ", my_number[7:]) #끝점은 생략가능
print("생년월일 역순 " + my_number[-7:]) #뒤에서부터 읽을 때에는 -를 붙여준다


##문자열 처리함수
python = "Python is Amazing"
print(python.lower()) #.lower()함수는 모두 소문자로 출력
print(python.upper())#.upper() 함수는 모두 대문자로 출력
print(python[0].isupper()) #대괄호 안에 index를 넣고
#.isupper()함수를 사용하면 해당위치에 대문자인지 판별해준다
print(python[0].islower())#대괄호 안에 index를 넣고
#.islower()함수를 사용하면 해당위치에 소문자인지 판별해준다
print(len(python)) #len(변수) 함수를 사용하면
# 그 변수의 길이를 출력해준다
print(python.replace("Python","Java")) #.replace(x,y)
#함수를 사용하면 x문자를 y문자로 치환해준다.
index = python.index("n") #.index(원하는 문자)는
#해당 변수에 그 문자가 몇 번째 위치에 있는지 알려준다
print(index)
index = python.index("n", index + 1)
#, index + a를 해놓으면 a만큼 뒤 index부터 찾기 시작함
print(index) #결과예상 : 15
print(python.find("Java")) #.find()함수도 .index()와
#비슷한 기능을 한다. 차이점 : -1이냐 error냐의 차이
print(python.count("n")) #.count(찾고싶은 것)
#.count() 함수는 괄호안에 있는 것이 몇번 나오는지 출력


#문자열포맷 방법 1
print("a" + "b") #형식지정방식 여러가지 중 택1
print("a", "b")
a = int(20)
print("나는 %d살 입니다." % a) # %d는 정수 
print("나는 %s을 좋아해요." % "파이썬")# %s는 문자열
print("Apple은 %c로 시작해요" % "A")
print("나는 %s색과 %s색을 좋아해요" % ("파랑", "빨강"))
# 형식지정자가 2개이상 나오면 꼭 괄호()로 묶어준다


#문자열포맷 방법2
print("나는 {}살 입니다.".format(20))
# 형식지정자 자리에 중괄호{}를 넣게 되면 뒤에
#.format()을 사용하여 괄호안에 값을 집어넣는다.
print("나는 {}색과 {}색을 좋아해요".format("파랑","빨강"))
# 2개 이상일 떄에는 똑같이 .format("","") 괄호로 묶어주고 콤마 사용한다


#문자열포맷 방법3
print("나는 {age}살이며, {color}색을 좋아해요.".format(age = 20, color = "red"))
# 형식지정자 부분의 중괄호{}와 .format까지는 동일하지만 괄호 안에 마치 변수를 설정하는 것처럼 적고 중괄호 속에 변수 이름을 입력하여 사용할 수도 있다.
name = "영준"
sc_num = "60181889"
# print("저의 이름은 {name}이고, 학번은 {sc_num}입니다.".format(name, sc_num))
# 위 코드가 error난 이유는 {}.format 형식에서는 그 줄에서 입력한 변수 값만 사용할 수 있기 떄문이다.


#문자열포맷 방법4
#★중요★ 위 방법처럼 중괄호 속에{} 값을 실제 선언한 지역변수로 사용하고 싶다면 print(f"") 따옴표 좌측 시작하는 곳에 f를 입력하면 된다.
print(f"저의 이름은 {name}이고, 학번은 {sc_num}입니다.")
#대신 뒤에.format() 함수를 사용하지 않음에 유의하자


#escape 문자
print("백문이 불여일견 \n백견이 불여일타")
#\n = 줄바꿈문자
print('"큰 따옴표를 출력하고 싶습니다."')
print("역슬래쉬 + \"를 통해 \"문자를 출력할 수 있음")
print("이번에는 \문자를 출력하고 싶습니다")
print("\문자를 출력하고 싶을 때에는 \\ 두개를 작성")
print("Red Apple\rPine")
#\r은 그 즉시 커서를 맨 앞으로 이동시켜준다
print("\\t는 tap입니다.")
print("\\b는 백스페이스 입니다")
print("박영\b준")


#Quiz 사이트별로 비밀번호를 만들어 주는 프로그램을 작성하시오 
#규칙1 : http:// 부분은 제외 => naver.com
#규칙2 : 처음 만나는 점(.)이후 부분은 제외 => naver
#규칙3 : 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e' + "!"로 구성
#예) 생성된 비밀번호 : nav51!
site_address = "https://youtube.com"
print("\n\n\n\n현재 사이트 주소 : %s \n" % site_address)
print("첫 번째 규칙은 https:// 부분을 제거하는 것입니다.")
del1 = site_address[8:]
print("첫 번째 규칙이 적용된 문자열 => " + del1)
print("\n두 번째 규칙은 .이후 부분을 제외하는 것입니다.")
del2 = site_address[8:15]
print("두 번째 규칙이 적용된 문자열 => " + del2)
print("\n세 번째 규칙은 남은 문자열의 첫 3자리 + 글자수 +글자 내의 \"e\" + \"!\"로 비밀번호를 생성하는 것입니다.")
del3 = site_address[8:11]
fin = (del3+str(len(del3))+str(del3.count("e"))+"!") 
#콤마, 가 아닌 +기호로 묶어줘야 문자열로 인식해서 딱 붙어 출력됨.
#str(len(변수)) = 그 변수의 길이 즉, 글자수를 출력해준다
#str(변수.count("x")) = 그 변수 속 x의 갯수를 출력해준다
print("3가지 규칙이 모두 적용된 비밀번호 : %s" % fin)
print("\n새로운 비밀번호가 생성되었습니다. 프로그램을 종료합니다.\n")
# find() or index()는 해당 문자의 index위치번호를 출력 find를써야 error 안 남

```
